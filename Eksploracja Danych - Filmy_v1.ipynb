```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea

df = pd.read_csv('movie_data.csv')

df = df.drop(columns='Rank')
df.head()
```

```python
df = df.sort_values("Revenue (Millions)", ascending=True)
df = df.dropna()
df.describe()
```

```python
movies_corr = df.corr()
f,ax = plt.subplots(figsize=(10, 10))
sea.heatmap(movies_corr, annot = True, linewidths = 0.1, fmt= '.2f', ax=ax )
plt.show()
```

```python
df_1 = df[['Rating', 'Revenue (Millions)']]
df_1.plot(x="Revenue (Millions)", y="Rating", kind='scatter')
```

```python
df_1 = df[['Metascore', 'Revenue (Millions)']]
df_1.plot(x="Revenue (Millions)", y="Metascore", kind='scatter')
```

```python
from sklearn.cluster import KMeans

df_1 = df[['Rating', 'Metascore']]
# df_1.plot(x="Rating", y="Revenue (Millions)", kind='scatter')

kmeans = KMeans(n_clusters=4, init='random', random_state=0).fit(df_1)

y_kmeans = kmeans.predict(df_1)

plt.scatter(df_1['Metascore'], df_1['Rating'], c=y_kmeans, s=50, cmap='viridis')
plt.xlabel("Metascore")
plt.ylabel("Rating")

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 1], centers[:, 0], c='black', s=50, alpha=0.5);
plt.show()
```

```python
from sklearn.cluster import KMeans

x1 = 'Revenue (Millions)'
x2 = "Rating"

df_1 = df[[x1, x2]]
df_1.plot(x=x1, y=x2, kind='scatter')

kmeans = KMeans(n_clusters=3, init='random', random_state=0).fit(df_1)

y_kmeans = kmeans.predict(df_1)

plt.scatter(df_1[x1], df_1[x2], c=y_kmeans, s=50, cmap='viridis')
plt.xlabel(x1)
plt.ylabel(x2)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=50, alpha=0.5);
plt.show()
```

## Klasyfikacja

Create dataframe with labels from clustering

```python
labeled_data = {'revenue': df_1['Revenue (Millions)'], 'rating': df_1['Rating'], 'labels': y_kmeans}

df_data_labels = pd.DataFrame(labeled_data)
df_data = df_data_labels[["revenue", "rating"]]
df_labels = df_data_labels[["labels"]]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

# split data to train, test dataset
x_train, x_test, y_train, y_test = train_test_split(df_data, df_labels.values.ravel(), test_size=0.3, random_state=42)

def plot_scores_for_knn(metric):
    neighbors_range = range(1, 15)
    scores = []
    for n in neighbors_range:
        knn = KNeighborsClassifier(n_neighbors=n, metric=metric)
        knn.fit(x_train, y_train)
        score = knn.score(x_test, y_test)
        scores.append(score)

    plt.figure()
    plt.plot(neighbors_range, scores)
    plt.title("Metric: " + metric)
    plt.xlabel("Number of neighbors")
    plt.ylabel("Score")
    plt.show()

plot_scores_for_knn('minkowski')
plot_scores_for_knn('manhattan')
plot_scores_for_knn('cosine')
```

### Drzewa decyzyjne

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier(random_state=42)
clf = clf.fit(x_train, y_train)

tree.plot_tree(clf, feature_names=['revenue', 'rating'], class_names=True, filled=True)
```
